# Example Graph

We consider a graph with 5 agents, also studied in Zhang and Yang's work, with the following Laplacian matrix:

$$
L = 
\begin{bmatrix}
1 & 0 & 0 & -1 & 0 \\
0 & 3 & -1 & -1 & -1 \\
0 & -1 & 3 & -1 & -1 \\
-1 & -1 & -1 & 3 & 0 \\
0 & -1 & -1 & 0 & 2 \\
\end{bmatrix}
$$

This case highlights the **numerical stability** of the iteration scheme in Eq. (9). After 400 iterations, $C^k \to 0$, and the resulting curve oscillates due to the limited precision of double-precision floating-point arithmetic. As a result, both the original and difference-based methods fail to accurately estimate the algebraic connectivity.


```python
from matplotlib import pyplot as plt
import numpy as np
from iter_draw import iter_draw_L

L = [
    [1, 0, 0, -1, 0],
    [0, 3, -1, -1, -1],
    [0, -1, 3, -1, -1],
    [-1, -1, -1, 3, 0],
    [0, -1, -1, 0, 2],
]
L = np.array(L)

f1, f2 = iter_draw_L(L, 2000,[-0.2, 1],True,1)
plt.show()
```
